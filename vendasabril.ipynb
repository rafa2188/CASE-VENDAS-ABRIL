```python
import pandas as pd
import numpy as np
import locale
from prettytable import PrettyTable
from termcolor import colored


tabela = pd.read_csv(r'Z:\Estudos Python_Sql_R_Powerbi_Excel\Base de Dados Estudos\VENDAS.csv')

display(tabela)
```

* Qual foi a media de vendas dos vendedores 

```python


# Definindo a localização como "pt_BR" para o Brasil
locale.setlocale(locale.LC_ALL, 'pt_BR')

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == 'April']

# Obtendo a mediana apenas para abril
mediana_abril = dados_abril["vendas totais"].median()

# Formatando a mediana em reais
mediana_formatada = locale.currency(mediana_abril, grouping=True, symbol=None)

# Imprimindo a mediana formatada
print("Mediana das vendas totais em abril:R$", mediana_formatada)
```

Qual o total de vendas dos vendedores 

```python


# Definindo a localização como "pt_BR" para o Brasil
locale.setlocale(locale.LC_ALL, 'pt_BR')

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == 'April']

# Obtendo a soma das vendas totais apenas para abril
soma_vendas_abril = dados_abril["vendas totais"].sum()

# Formatando a soma das vendas em reais
soma_vendas_formatada = locale.currency(soma_vendas_abril, grouping=True, symbol=None)

# Imprimindo a soma das vendas formatada
print("Soma das vendas totais em abril:R$", soma_vendas_formatada)
```

* Qual o top 10 vendedores em abril

* Qual marca mais vendeu 

```python
from termcolor import colored

mes_de_interesse = 'April'

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == mes_de_interesse]

# Contar a frequência das marcas em abril
brand_counts = dados_abril['marca'].value_counts()

# Obter a marca mais vendida em abril
most_sold_brand = brand_counts.index[0]

# Filtrar os dados apenas para a marca mais vendida em abril
dados_marca_mais_vendida = tabela[(tabela['mês'] == mes_de_interesse) & (tabela['marca'] == most_sold_brand)]

# Calcular a quantidade total vendida da marca mais vendida em abril
quantidade_total_vendida = dados_marca_mais_vendida['quantidade vendida'].sum()

# Criar a string com as informações coloridas em vermelho
informacoes = "A marca mais vendida em abril é: " + colored(most_sold_brand, 'red') + ". Quantidade total vendida: " + colored(str(quantidade_total_vendida), 'red')

# Imprimir as informações na mesma linha
print(informacoes)




```

* O produto mais vendido em abril é

```python
from termcolor import colored

mes_de_interesse = 'April'

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == mes_de_interesse]

# Contar a frequência dos produtos em abril
product_counts = dados_abril['nome do produto'].value_counts()

# Obter o produto mais vendido em abril
most_sold_product = product_counts.index[0]

# Filtrar os dados apenas para o produto mais vendido em abril
dados_produto_mais_vendido = dados_abril[dados_abril['nome do produto'] == most_sold_product]

# Calcular a quantidade total do produto vendido em abril
quantidade_total_vendida = dados_produto_mais_vendido['quantidade vendida'].sum()

# Criar a string com as informações coloridas em vermelho
informacoes = "O produto mais vendido em abril é: " + colored(most_sold_product, 'red') + ". Quantidade total vendida: " + colored(str(quantidade_total_vendida), 'red')

# Imprimir as informações na mesma linha
print(informacoes)


```

* Maior e Menor quantidade de produto vendida em abril por unidade :

```python


mes_de_interesse = 'April'

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == mes_de_interesse]

# Encontrar a maior quantidade de produto vendida em abril e seus detalhes
maior_quantidade = dados_abril['quantidade vendida'].max()
maior_quantidade_produto = dados_abril[dados_abril['quantidade vendida'] == maior_quantidade]
marca_maior_quantidade = maior_quantidade_produto['marca'].iloc[0]
tipo_produto_maior_quantidade = maior_quantidade_produto['nome do produto'].iloc[0]
unidades_vendidas_maior_quantidade = maior_quantidade_produto['quantidade vendida'].iloc[0]
total_maior_quantidade = maior_quantidade_produto['quantidade vendida'].sum()

# Encontrar a menor quantidade de produto vendida em abril e seus detalhes
menor_quantidade = dados_abril['quantidade vendida'].min()
menor_quantidade_produto = dados_abril[dados_abril['quantidade vendida'] == menor_quantidade]
marca_menor_quantidade = menor_quantidade_produto['marca'].iloc[0]
tipo_produto_menor_quantidade = menor_quantidade_produto['nome do produto'].iloc[0]
unidades_vendidas_menor_quantidade = menor_quantidade_produto['quantidade vendida'].iloc[0]
total_menor_quantidade = menor_quantidade_produto['quantidade vendida'].sum()

# Criar a tabela
tabela = PrettyTable()
tabela.field_names = ["", "Marca", "Tipo de Produto", "Unidades Vendidas", "Quantidade Total Vendida"]
tabela.add_row(["Maior quantidade de um único produto vendida em abril", marca_maior_quantidade,
                tipo_produto_maior_quantidade, colored(str(unidades_vendidas_maior_quantidade), 'red'),
                colored(str(total_maior_quantidade), 'red')])
tabela.add_row(["Menor quantidade de produto vendida em abril", marca_menor_quantidade,
                tipo_produto_menor_quantidade, colored(str(unidades_vendidas_menor_quantidade), 'red'),
                colored(str(total_menor_quantidade), 'red')])

# Imprimir a tabela
print(tabela)



```

```python

mes_de_interesse = 'April'

# Filtrar os dados apenas para o mês de abril
dados_abril = tabela[tabela['mês'] == mes_de_interesse]

# Agrupar por vendedor e calcular a quantidade de vendas
vendas_por_vendedor = dados_abril.groupby('nome do vendedor')['quantidade vendida'].sum().reset_index()

# Ordenar os vendedores por quantidade de vendas em ordem decrescente
vendas_por_vendedor = vendas_por_vendedor.sort_values('quantidade vendida', ascending=False)

# Criar uma função para aplicar a formatação de cores aos valores
def colorizar_valor(valor):
    valor_sem_zeros = str(valor).rstrip('0').rstrip('.')  # Remover os zeros à direita e o ponto decimal, se necessário
    if valor_sem_zeros in vendas_por_vendedor['quantidade vendida'].nlargest(22).astype(str).str.rstrip('0').str.rstrip('.').values:
        return 'background-color: green'
    elif valor_sem_zeros in vendas_por_vendedor['quantidade vendida'].nsmallest(22).astype(str).str.rstrip('0').str.rstrip('.').values:
        return 'background-color: red'
    else:
        return 'background-color: yellow'

# Aplicar a formatação de cores aos valores da tabela
vendas_por_vendedor_styled = vendas_por_vendedor.style.applymap(colorizar_valor, subset='quantidade vendida')

# Configurar a exibição completa dos dados no DataFrame
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

# Exibir a tabela completa de vendas por vendedor em abril com cores
print("Vendas por vendedor em abril:")
display(vendas_por_vendedor_styled)




```
